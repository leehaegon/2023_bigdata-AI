```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 데이터 준비
data = {
    'address': [
        '서울특별시 중구 을지로 39길 29',
        '충청남도 천안시 동남구 천안대로 357',
        '광주광역시 동구 필문대로 365',
        '경상남도 양산시 물금읍 금오로 20',
        '대구시 중구 달성로 56',
        '서울특별시 성동구 고산자로 270',
        '부산 부산진구 시민공원로 73',
        '인천광역시 연수구 경원대로 526',
        '광주광역시 서구 금화로 278',
        '대전광역시 유성구 유성대로 978'
    ],
    'createdAt': [
        '2021-03-03 07:00:52',
        '2021-03-03 07:00:52',
        '2021-03-03 07:00:52',
        '2021-03-03 07:00:53',
        '2021-03-03 07:00:53',
        '2021-03-15 00:03:43',
        '2021-03-15 00:03:43',
        '2021-03-15 00:03:43',
        '2021-03-15 00:03:43',
        '2021-03-15 00:03:44'
    ]
}

df = pd.DataFrame(data)

# 주소 막대 그래프
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['address'])

# x축 레이블 설정
plt.xticks(df.index, df['address'], rotation=90)

# 그래프 제목과 축 이름 설정
plt.title('Address Bar Chart')
plt.xlabel('Index')
plt.ylabel('Address')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

```python
# 시계열 그래프
df['createdAt'] = pd.to_datetime(df['createdAt'])  # 문자열을 날짜/시간 형식으로 변환

plt.figure(figsize=(10, 6))
plt.plot(df['createdAt'], df.index, marker='o')

# x축 형식 설정
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))

# 그래프 제목과 축 이름 설정
plt.title('Time Series Chart')
plt.xlabel('Date')
plt.ylabel('Index')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```
